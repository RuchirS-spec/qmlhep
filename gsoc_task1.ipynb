```python
!pip install pennylane
import pennylane as qml
from pennylane import numpy as np
```

```python
dev = qml.device("default.qubit", wires=5) #set device, set 5 qubits

@qml.qnode(dev)
def circuit1():

    for i in range(5):
      qml.Hadamard(wires=i) #b) hadamard on every qubit

    qml.CNOT(wires=[0, 1])  #c) CNOT operation on (0, 1),(1,2),(2,3),(3,4)
    qml.CNOT(wires=[1, 2])
    qml.CNOT(wires=[2, 3])
    qml.CNOT(wires=[3, 4])

    qml.SWAP(wires=[0, 4]) #d) SWAP (0, 4)

    qml.RX(np.pi/2, wires=3) #e) Rotate X with pi/2 on qubit(3)

    return qml.state()
```

```python
print("Circuit:")  #f) plot the circuit
print(qml.draw(circuit1)())
```

```python
dev = qml.device("default.qubit", wires=5) #set device, set 5 qubits

@qml.qnode(dev)
def circuit2():

    qml.Hadamard(wires=0) #a) applied a hadamard gate to the first qubit

    qml.RX(np.pi/3, wires=1)#b) rotate the second qubit by pi/3 around X

    qml.Hadamard(wires=2) #c) applied a hadamard gate to the third and fourth qubits
    qml.Hadamard(wires=3)

    #seting qubit 5 as ancilla
    qml.Hadamard(wires=4)
    # swap test between qubit 1 and 2
    qml.CSWAP(wires=[4, 0, 1])
    # swap test between qubit 3 and 4
    qml.CSWAP(wires=[4, 2, 3])
    qml.Hadamard(wires=4)

    return qml.probs(wires=4)
```

```python
print("Circuit for Swap Test:")
print(qml.draw(circuit2)())
```
